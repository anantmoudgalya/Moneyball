### From above graph we can observe most teams have made the playoffs after winning 95+ games

* So how many wins would the A's need to make the playoffs? 95+.
* The began recruiting players in terms of runs scored, on-base percentage and slugging percentage
* To determine the exact OBPs of the players needed, we can obtain that from Run Differential
* Run Differential = Runs Scored - Runs Allowed

```python
RD = df["RS"]-df["RA"]
df["RD"] = RD
```

```python
fig = plt.figure(num=None, figsize=(10, 10), dpi=80)
plt.grid(True, linewidth=0.5, linestyle='--')
sns.scatterplot(x="RD", y="W", hue="Playoffs", data=df, palette=["black","blue"])
plt.xlabel("Run Differential")
plt.ylabel("Wins")
plt.show()
```

* We can see that the required run differential is >100 and <200

### Using past data, we can build a simple regression model for Wins vs Run Differential

```python
df.columns
```

```python
x = df["RD"]
y = df["W"]
print(len(x), len(y))
```

```python
#regression model
from sklearn.linear_model import LinearRegression
from sklearn import metrics

lr = LinearRegression()
reg = lr.fit(x.values.reshape(-1,1),y.values.reshape(-1,1))
```

```python
print(reg.coef_, reg.intercept_)
```

* Linear Regression equation can be written as,
        
        W = 80.881 + (0.105)*RD
        
* Substituting, W = 95, we can calculate Run Differential as,
        
         RD = 134.46 ~ 134.

```python
fig = plt.figure(num=None, figsize=(10, 10), dpi=80)
plt.grid(True, linewidth=0.5, linestyle='--')
sns.scatterplot(x="RD", y="W", hue="Playoffs", data=df, palette=["black","blue"])
plt.axvline(x = 134, linestyle="dashed", color = 'black')
plt.axhline(y = 95, linestyle="dashed", color = 'black')
plt.text(x=139, y=115, s="Made Playoffs")
plt.xlabel("Run Differential")
plt.ylabel("Wins")
plt.show()
```

* Now the question to answer was, how do score 134 runs more than we allow?

* To determine this, we can find out which stats are most significant towards run scoring

```python
bat_data = df[["RS", "OBP", "SLG", "BA"]]
bat_data.head()
```

```python
fig, ax = plt.subplots(figsize = (10,10))
sns.heatmap(bat_data.corr(), ax =ax, annot = True, linewidths=0.05,fmt='.2f',cmap="magma")
```

* We can see that OBP (On-base percentage) and SLG (Slugging Rate) affect Runs Scored(RS) more than BA
* Hence, we can build a simple regression model for RS based on OBP and SLG
* Sidenote: During the time Beane and DePodesta were performing these calculations, traditional major league scouts placed a heavy emphasis on batting average (BA). Before looking at detailed statistics, it would seem like BA would be the most directly correlated with scoring runs, but that is not the case. Traditional scouts had no clue that they were looking for the wrong attributes all along. That blissful ignorance was one of the main reasons Beane and DePodesta were able to gain such an advantage.

```python
lr  = LinearRegression()

x_bat = bat_data[["OBP","SLG"]]
y_bat = bat_data["RS"]

reg = lr.fit(x_bat,y_bat)
```

```python
print(reg.coef_, reg.intercept_)
```

* So, now the regression model can be written as
        
        RS = 2737.8*OBP + 1584.9*SLG - 804.63
        
* Using the OOBP and OSLG for the A's from the 2001 season, DePodesta assumed that the numbers would be roughly the same for the 2002 season. He used a regression model to predict how many runs the As would allow in 2002

```python
ra_data = df[["RA","OOBP","OSLG"]]
print(ra_data.isnull().sum())
ra_data = ra_data.iloc[:90] #Accomodating for null values
```

```python
ra_data.head()
```

```python
lr = LinearRegression()

x_ra = ra_data[["OOBP","OSLG"]]
y_ra = ra_data["RA"]

reg = lr.fit(x_ra, y_ra)
```

```python
print(reg.coef_, reg.intercept_)
```

* So, now the Runs Allowed (RA) regression model can be written as
        
        RA = 2913.6*OOBP + 1514.29*OSLG - 837.38

## Final Calculations

* The OOBP and OSLG for As in 2001 were 0.308 and 0.380 respectively

    RA = 2913.6(0.308) + 1514.29*(0.380) - 837.38
       = 635.43 ~ 635.
    
* Beane and DePodesta would have to build a team to score (635+134) ~ 770 runs in 2002 to make the playoffs. Since the scouts were valuing players only in terms of Batting Average(BA), they were able to create a highly effective roster under their marginalised salary budget.

* Now, we can observe that the As had the following statistics in 2002
    
    OBP = 0.339
    SLG = 0.432
    OOBP = 0.315
    OSLG = 0.384
    
* Plugging this in the RA equation, we get,
    
    RA = 2913.6(0.315) + 1514.29*(0.384) - 837.38
       = 661.89 ~ 662
       
    The As allowed 654 runs in 2002

* Plugging OBP and SLG in RS equation, we get,

    RS = 2737.8*(0.339) + 1584.9*(0.432) - 804.63
       = 808.1 ~ 808.
   
    The As scored 800 runs in 2002.
    
* Calculated Run Differential = RS - RA = 808 - 662 = 146.
* Actual Run Differential = RS - RA = 800 - 654 = 146.
* Therefore, we can see that our equations are actually pretty accurate.

* Plugging RD value in our Wins equation,

    W = 80.881 + (0.105)*146
      = 96.
      
    The As won 103 games that season.
    
* We can see that all the values satisfy and surpass the thresholds set in our equations, once again proving that finding the real value in data can be significantly beneficial.

This kind of smart, statistical analyisis by DePodesta and excellent roster management by Billy Beane led the Oakland As to 103 wins, and they made the playoffs, ranking first in the regular season with a 41mil payroll while the Yankees' payroll topped 125mil the same year.
